```python
import serial
import time
import numpy as np
import matplotlib.pyplot as plt
```

# Program(1) : Recieving data from UltraSonic and display it.

```python
serialcomm = serial.Serial('/dev/ttyUSB0','9600')
serialcomm.timeout = 1

while True:
    

    time.sleep(0.5)

    l = serialcomm.readline().decode('ascii').split(","); 
    if len(l)== 2:
        print("Distance in meter = " + l[0])
        print("Distance in inch = " + l[1])
    else:
        print("Distance in inch = " + l[0])


serialcomm.close()
```

# Program(2) : Recieving data from UltraSonic and visulize it.

```python
serialcomm = serial.Serial('/dev/ttyUSB0','9600')
serialcomm.timeout = 1

plt.figure()
plt.ion()
plt.show()


data = np.array([])
while True:
    

    time.sleep(0.5)

    l = serialcomm.readline().decode('ascii').split(",")
    data = np.append(data,float(l[0]))
    plt.plot(data)
    plt.title("Visualizing distance from UltraSonic sensor")
    plt.xlabel("time 's'")
    plt.ylabel("Distance in 'mm'")
    plt.pause(.01)
                     


serialcomm.close()
```

# Program(3): Calculate the speed and modify the file

```python
def calc_speed(l):
    speed = 0
    for i in l :
        speed += float(i)**2
    return speed
```

```python
fh = open("python_test_file3.txt",'r')

v = []
for line in fh:
    if "velocity" in line :
    
        l = line.split(',')
        #access the last three elements which represent the velocities
        x = l[-1:-4:-1] 
        #check that x is not empty
        if '        '  not in x:
            
            speed = calc_speed(x)

            line = "{},speed = {:0.3f} \n".format(line[0:172], speed)
            v.append(line)
            print(line)
        
        else:
           
            line = "{},speed = {:0.3f} \n".format(line[0:172], 0)
            v.append(line)
            print(line)
            

            
fh.close() 



#fname = input("Etern file name: ")
fh = open("python_test_file3.txt",'w')

for i in range(len(v)):
    
    fh.write(v[i])
    
fh.close() 
```

# Program(4) : Get All Arduino IDs from a file 

```python
f = open("FileWith2ArduinoConnected.txt",'r')

v = []

for line in f:
    
    if "E: ID_SERIAL=Arduino" in line :
        v.append(line)
    

x = list(set([l[43:-1] for l in v]))
            
for i in range(len(x)):
    print(f"Arduino ({i+1}) ID Number : {x[i]}")
            
f.close()
```
