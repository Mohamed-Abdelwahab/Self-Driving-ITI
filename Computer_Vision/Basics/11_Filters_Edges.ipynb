## Required 

Implement and apply the following filters on carkyo image:
- Scharr.
- Prewitt.
- Roberts.

## Criteria

- Filter must be in both X-Dir and Y-Dir.
- For each filter, 3 outputs are required (x, y, mag).
- All filters must be implemented using cv2.Filter2D.

```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

```python
img = cv2.imread('carkyo.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

## Scharr Filter 

```python
# X-Direction
Scharr_x = np.array(
    [
        [3,0,-3],
        [10,0,-10],
        [3,0,-3],
    ]
, dtype=np.float32)

out1 = cv2.filter2D(gray, -1, Scharr_x)
plt.imshow(out1, cmap='gray');
```

```python
# Y-Direction
Scharr_y = np.array(
    [
        [3,10,3],
        [0,0,0],
        [-3,-10,-3],
    ]
, dtype=np.float32)
out2 = cv2.filter2D(gray, -1, Scharr_y)
plt.imshow(out2, cmap='gray');
```

```python
# Magnitude of two images
out3 = (out1**2 + out2**2)**0.5
plt.imshow(out3, cmap='gray');
```

## Prewitt Filter 

```python
# X-Direction
Prewitt_x = np.array(
    [
        [-1,0,1],
        [-1,0,1],
        [-1,0,1],
    ]
, dtype=np.float32)

out4 = cv2.filter2D(gray, -1, Prewitt_x)
plt.imshow(out4, cmap='gray');
```

```python
# Y-Direction
Prewitt_y = np.array(
    [
        [1,1,1],
        [0,0,0],
        [-1,-1,-1],
    ]
, dtype=np.float32)

out5 = cv2.filter2D(gray, -1, Prewitt_y)
plt.imshow(out5, cmap='gray');
```

```python
# Magnitude of two images
out6 = (out4**2 + out5**2)**0.5
plt.imshow(out6, cmap='gray');
```

## Roberts Filter

```python
# X-Direction

Roberts_x = np.array(
    [
        [1,0],
        [0,-1]
    ]
, dtype=np.float32)

out7 = cv2.filter2D(gray, -1, Roberts_x)
plt.imshow(out7, cmap='gray');
```

```python
# Y-Direction
Roberts_y = np.array(
    [
        [0,1],
        [-1,0]
    ]
, dtype=np.float32)

out8 = cv2.filter2D(gray, -1, Roberts_y)
plt.imshow(out8, cmap='gray');
```

```python
# Magnitude of two images
out9 = (out7**2 + out8**2)**0.5
plt.imshow(out9, cmap='gray');
```
