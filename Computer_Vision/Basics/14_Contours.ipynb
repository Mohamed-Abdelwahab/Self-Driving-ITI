```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# read image and convert to gray
img = cv2.imread('Files/Contours/lab.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(25,10))
plt.imshow(gray,cmap="gray")
```

```python
# Apply color thrsholding
_, th = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
# finding contours i a binaty image                       

contours, _ = cv2.findContours(th, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
len(contours)
cv2.drawContours(img, contours,0, (255, 255, 255), -1)
cv2.drawContours(img, contours,5, (255, 255, 255), -1)
cv2.drawContours(img, contours,6, (255, 255, 255), -1)
cv2.drawContours(img, contours,3, (0, 0, 255), -1)
cv2.drawContours(img, contours,4, (255, 0, 0), -1)
plt.figure(figsize=(25,10))
plt.imshow(img)
```
