## Normal Equalizer 

- Computing a histogram of image pixel intensities
- Evenly spreading out and distributing the most frequent pixel values
- Giving a linear trend to the cumulative distribution function 
- The result of applying histogram equalization is an image with higher global contrast which makes bixels too bright or too dark (contrast problem)


```python
equalized = cv2.equalizeHist(gray)
plt.imshow(equalized,cmap='gray')
```

## What it 'CLAHE'?

- LAHE is a variant of Adaptive histogram equalization (AHE) which takes care of over-amplification of the contrast. CLAHE operates on small regions in the image, called tiles, rather than the entire image. The neighboring tiles are then combined using bilinear interpolation to remove the artificial boundaries.
This algorithm can be applied to improve the contrast of images.

## CLAHE Histogram Eqalization 

- solve the problem of contrast by limiting it to specific thresholding value
- divide image into (8,8) windows and apply bi-linear interpolation 

```python
clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(8, 8))
equalized2 = clahe.apply(gray)
plt.imshow(equalized2,cmap='gray')
```
