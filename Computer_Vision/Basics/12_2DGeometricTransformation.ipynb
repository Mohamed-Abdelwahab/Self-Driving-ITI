```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
img1 = mpimg.imread('Files/Affine/perspective02.jpg')

plt.imshow(img1)
plt.show()
```

```python
rows,cols = img1.shape[:2]
```

```python
%matplotlib notebook

pts1 = np.float32(
    
    [
        [580,  350], [700, 350],
        [1279, 718], [0, 718]
    ])

    # step(5) : destination points
pts2 = np.float32(
    [
        [580, 20], [700,20],
        [700, 700], [580, 700]
    ]
)

    # step(6) : find the matrix the transform you from source -> destination 
M = cv2.getPerspectiveTransform(pts1 ,pts2)
    # step (7) : apply transformation to the whole image 
dst = cv2.warpPerspective(img1, M, (cols,rows))

plt.imshow(dst)
plt.show()
```

```python
%matplotlib notebook
img2 = mpimg.imread('Files/Affine/perspective01.png')

plt.imshow(img2)
plt.show()
```

```python
rows2,cols2 = img2.shape[:2]
```

```python
%matplotlib notebook

src = np.float32(
    
    [
        [200,200], [625, 200],
        [3, 255], [775, 255]
    ])

    # step(5) : destination points
dst = np.float32(
    [
        [200, 200], [625,200],
        [200, 500], [625, 500]
    ]
)

    # step(6) : find the matrix the transform you from source -> destination 
M2 = cv2.getPerspectiveTransform(src ,dst)
    # step (7) : apply transformation to the whole image 
dst2 = cv2.warpPerspective(img2, M2, (cols2,rows2))

plt.imshow(dst2)
plt.show()
```
